```python
#@title recursion
def proses_data(vitamins_data, index=0):
    if index >= len(vitamins_data):
        return
    kelompok_umur = vitamins_data.iloc[index]["Life-Stage Group"]
    print(f"Proses data untuk {kelompok_umur}:")
    for key, value in vitamins_data.iloc[index].items():
        if key != "index" and key != "Life-Stage Group":
            print(f"  {key}: {value} {get_satuan(key)}")
    proses_data(vitamins_data, index + 1)
def get_satuan(key):
    if key.endswith(" (µg/d)"):
        return "µg/d"
    elif key.endswith(" (mg/d)"):
        return "mg/d"
    else:
        return ""
proses_data(vitamins_data)
```

```python
#@title first class
# Definisikan fungsi untuk menghitung kebutuhan vitamin harian
def hitung_kebutuhan_vitamin(life_stage_group, vitamin):
    # Definisikan dictionary untuk menyimpan kebutuhan vitamin harian
    kebutuhan_vitamin = {
        "Children 1-3 y": {
            "Vitamin A": 300,
            "Vitamin C": 15,
            "Vitamin D": 15,
            "Vitamin E": 6,
            "Vitamin K": 30.0,
            "Thiamin": 0.5,
            "Riboflavin": 0.5,
            "Niacin": 6,
            "Vitamin B6": 0.5,
            "Folate": 150,
            "Vitamin B12": 0.9,
            "Pantothenic Acid": 2.0,
            "Biotin": 8,
            "Choline": 200
        },
        "Infants 0-6 mo": {
            "Vitamin A": 400,
            "Vitamin C": 40,
            "Vitamin D": 10,
            "Vitamin E": 4,
            "Vitamin K": 2.0,
            "Thiamin": 0.2,
            "Riboflavin": 0.3,
            "Niacin": 2,
            "Vitamin B6": 0.1,
            "Folate": 65,
            "Vitamin B12": 0.4,
            "Pantothenic Acid": 1.7,
            "Biotin": 5,
            "Choline": 125
        },
    }

    # Cek apakah life stage group dan vitamin yang diinginkan ada dalam dictionary
    if life_stage_group in kebutuhan_vitamin and vitamin in kebutuhan_vitamin[life_stage_group]:
        return kebutuhan_vitamin[life_stage_group][vitamin]
    else:
        return "Kebutuhan vitamin tidak ditemukan"

# Definisikan fungsi first class untuk menghitung kebutuhan vitamin harian
def hitung_kebutuhan_vitamin_harian(life_stage_group):
    def hitung_kebutuhan_vitamin_(vitamin):
        return hitung_kebutuhan_vitamin(life_stage_group, vitamin)
    return hitung_kebutuhan_vitamin_

# Contoh penggunaan fungsi first class
hitung_kebutuhan_vitamin_children_1_3_y = hitung_kebutuhan_vitamin_harian("Children 1-3 y")
print(hitung_kebutuhan_vitamin_children_1_3_y("Vitamin A"))
print(hitung_kebutuhan_vitamin_children_1_3_y("Vitamin C"))
```
